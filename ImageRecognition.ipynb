```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt

folders = [r'C:\Users\Loga lv\Downloads\strawberry', r'C:\Users\Loga lv\Downloads\crab_apple', r'C:\Users\Loga lv\Downloads\blueberry']

file_paths_with_class = []

for folder in folders:
    files = os.listdir(folder)
    file_paths_with_class.extend([(os.path.join(folder, file), os.path.basename(folder).split('_')[-1]) for file in files])

random.shuffle(file_paths_with_class)

random_images = file_paths_with_class[:10]

for image_path, class_name in random_images:
    img = Image.open(image_path)
    plt.imshow(img)
    plt.title(f"Class: {class_name}")
    plt.axis('off')
    plt.show()
```
